## 1. Crear una columna basado en 1 condicion: np.where()

```python
import numpy as np
```

```python
# Crear un array basado en nivel de precio (price_tier)
np.where(df_laptops['Price_euros'] > 2000, 'Caro', 'Barato')
```

```python
# anadir nueva columna
df_laptops['price_category'] = np.where(df_laptops['Price_euros'] > 2000, 'Caro', 'Barato')
```

```python
# Mostrar 5 primeras columnas
df_laptops.head(5)
```

```python
# contar los valores en columna price_tier
df_laptops.value_counts('price_category')
```
