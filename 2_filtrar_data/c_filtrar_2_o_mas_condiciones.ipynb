### 3 Crear columna condicional con 2 o mas opciones np.select()

```python
import numpy as np
```

```python
# crear un array basado en multiple niveles de precio (+2 opciones)

# crear "conditions" (condiciones) y "values" (valores)

conditions = [
    df_laptops['Price_euros'] > 3000,
    (df_laptops['Price_euros'] > 2000) & (df_laptops['Price_euros'] <= 3000),
    (df_laptops['Price_euros'] > 800) & (df_laptops['Price_euros'] <= 2000),
    df_laptops['Price_euros'] <= 800
]

values = [
    'Muy caro',
    'Caro',
    'Barato',
    'Muy barato'
]

# >3000, >2000, >800, <=800
```

```python
# añadir a una columna
np.select(conditions, values, default='Desconocido')
```

```python
# mostrar dataframe
df_laptops['price_category'] = np.select(conditions, values, default='Desconocido')
df_laptops
```

```python
# contar los valores en columna price_tier
df_laptops.value_counts('price_category')
```

##### 3.1 Ejercicio

```python
# crear un array basado en multiple tamaños de pantalla (screen size +2 opciones)
# Muy grande > 16, Grade > 14, Pequeño > 12, Muy Pequeño < 12

# Crear Condiciones y valores

conditions_screen = [
    df_laptops['Inches'] > 16,
    (df_laptops['Inches'] > 14) & (df_laptops['Inches'] <= 16),
    (df_laptops['Inches'] > 12) & (df_laptops['Inches'] <= 14),
    df_laptops['Inches'] <= 12
]

values_screen = [
    'Muy grande',
    'Grande',
    'Pequeño',
    'Muy pequeño'
]
```

```python
screen_size = np.select(conditions_screen, values_screen, default='Desconocido')
```

```python
# Añadir nueva columna
df_laptops['Screen size'] = screen_size
df_laptops
```

```python
df_laptops[df_laptops['Screen size']=='Muy pequeño']
```

```python
# Contar los valores en columna Screen_size
df_laptops.value_counts('Screen size')
```
