```python
import pandas as pd
```

```python
# Leer archivo csv
df_futbol = pd.read_csv('players_20_new.csv')
df_futbol.set_index('short_name', inplace=True)
df_futbol
```

### 1. Seleccionar con un valor



```python
# Obtener el tamano de L. Messi
df_futbol.iloc[0, 3]
```

```python
# Obtener el peso de Cristiano Ronaldo

df_futbol.iloc[1, 4]
```

```python
# Obtener todas las filas dentro de la columna 'height_cm'
df_futbol.iloc[:, 3]
```

```python
# Obtener toda las columnas de L. Messi
df_futbol.iloc[0, :]
```

### 2. Seleccionar con una lista de valores

```python
# Obtener data sobre L. Messi y Cristiano 
df_futbol.iloc[[0,1]]
```

```python
# Obtener el tamaño de L. Messi y Cristiano Ronaldo
df_futbol.iloc[[0,1], 3]
```

```python
# Obtener el tamaño y peso de L. Messi
df_futbol.iloc[0,[3,4]]
```

```python
# Obtener el tamaño y peso de L. Messi y Cristiano Ronaldo
df_futbol.iloc[[0,1],[3,4]]
```

### 3. Seleccionar un Rango de data con un Slice (Rebanada)

```python
# Hacer slice a nombre de columnas
players = [0,1]

df_futbol.iloc[players, 1:7]
```

```python
# Hacer slice a los nombre de los index
columns = [1, 2, 3, 4]

# Obtener los nombres de jugadores topl y top10

df_futbol.iloc[:10, columns]
```

### 4. Seleccionar con condiciones

```python
# una condicion: seleccionar jugadores con tamano arriba de 180 cm
df_futbol.iloc[list(df_futbol['height_cm']>180), columns]
```
