```python
import pandas as pd
```

```python
df_futbol = pd.read_csv('players_20_new.csv')
df_futbol.set_index('short_name', inplace=True)
df_futbol
```

### 1. Fijar valor a una celda

```python
# Actualizar altura de L. Messi
df_futbol.loc['L. Messi', 'height_cm'] = 175
```

```python
# Mostrar DF
df_futbol
```

### 3 Fijar valor de toda la columna

```python
# Fijar tamano de 190 a todos los jugadores
df_futbol.loc[:,'height_cm'] = round(df_futbol.loc[:,'height_cm'].mean())
```

```python
# Mostrar DF
df_futbol
```

### 3 Fijar valor a toda la fila

```python
# Obtener todas las columnas que corresponden al jugador rankeado utimo al jurgo fifa
df_futbol.iloc[-1,:]
```

```python
# Fijar valor nulo a todas las columnas que corresponden al jugador rankeado
import numpy as np

df_futbol.iloc[-1,:] = np.nan
```

```python
# Mostrar DF
df_futbol
```

### 4 Fijas valor a multiples celdas

```python
# Fijar valor a todos los items que hacen match con la lita de nombres


df_futbol.loc[['L. Messi', 'Cristiano Ronaldo'], ['height_cm']] = 175
```

```python
# Mostrar DF
df_futbol
```

### 5 Fijar valor para filas que cumplen una condicion

```python
# Fijar valor para filas que cumplen una condicion
columns = ['age', 'dob', 'height_cm', 'weight_kg']

df_futbol.loc[df_futbol['height_cm'] > 180, columns] = 0
```

```python
# Mostrar DF
df_futbol
```
