## 2. Estrategia Split-Apply-Combine

```python
df_cars_sales.value_counts('Vehicle_type')

#Split

car_filter = df_cars_sales['Vehicle_type'] == 'Car'
passenger_filter = df_cars_sales['Vehicle_type'] == 'Passenger'
```

```python
# Apply

car_average = df_cars_sales[car_filter]['Sales_in_thousands'].mean()
passenger_average = df_cars_sales[passenger_filter]['Sales_in_thousands'].mean()
```

```python
# Combine

pd.DataFrame({'Vehicule_Type':['Car', 'Passenger'],
              'Sales_in_thousands': [car_average, passenger_average]})
```

## 5. Lambda

```python
df_cars_sales.groupby('Manufacturer').sum().head()
```

```python
df_cars_sales.groupby('Manufacturer').sum().apply(lambda x:x*1000)[df_cars_sales.select_dtypes(include='number').columns][['Sales_in_thousands', 'Price_in_thousands']]
```

```python
# Restar valores de la media a cada grupo
df_cars_sales.groupby('Manufacturer')[df_cars_sales.select_dtypes(include='number').columns].apply(lambda x:x-x.mean())[:'Acura']
```
