```python
import pandas as pd
```

```python
df_movies = pd.read_csv("IMDb movies.csv", low_memory=False)
df_ratings   = pd.read_csv("IMDb ratings.csv")
```

```python
df_ratings.columns
```

```python
df_movies = df_movies[['imdb_title_id', 'title', 'year', 'genre', 'country']]
df_ratings = df_ratings[['imdb_title_id', 'total_votes', 'mean_vote']]
```

```python
df_movies
```

```python
df_ratings
```

## 1. CONCATENAR (concat())

#### 1.1 Concatenar Verticalmente

```python
df1 = pd.DataFrame({
    'id': ['A', 'B', 'C', 'D'],
    'age': [30, 23, 25, 22]
})
df2 = pd.DataFrame({
    'id': ['E', 'F', 'G', 'H'],
    'age': [40, 21, 19, 24]
})
```

```python
pd.concat([df1, df2], axis=0, ignore_index=True, sort=True)
```

##### 1.1.1 Ejercicio 1

```python
# Extraer 50% de muestra del dataframe original (df_movies)
df_sample = df_movies.sample(frac=0.5)
```

```python
# shape
print(df_sample.shape, df_movies.shape)
```

```python
# concatenar verticalmente df_movie y df_samples
df_concat_vert = pd.concat([df_movies, df_sample], axis=0)
```

```python
df_concat_vert.shape
```

#### 1.2 Concatenar Horizontalmente

```python
df3 = pd.DataFrame({
    'job': ['Doctor', 'Statistician', 'Accountant', 'Developer']
})

pd.concat([df1, df3], axis=1, sort=True)
```

##### 1.2.1 Ejercicio 2

```python
# Shape
print(df_movies.shape, df_ratings.shape)
```

```python
# Concatenar horizontal df_movies y df_ratings usando la columna 'imdb_title_id'
df_movies.set_index('imdb_title_id', inplace=True)
df_ratings.set_index('imdb_title_id', inplace=True)
```

```python
df_movies_ratings = pd.concat([df_movies, df_ratings], axis=1)
df_movies_ratings
```

```python
# Shape
df_movies_ratings.shape
```

## 2. JOINS (merge())

#### 2.1 Inner Join

```python
df3.insert(loc=0, column='id', value=['C', 'D', 'E','F'])
df3
```

```python
df_inner_join = df1.merge(df3, on='id', how='inner')
df_inner_join
```

```python
print(df1.shape, df3.shape, df_inner_join.shape)
```

##### 2.1.1 Ejercicio 

```python
# merge df_movies and df_ratings (inner join)
df_movies_inner = df_movies.merge(df_ratings, on='imdb_title_id', how='inner')
df_movies_inner
```

```python
df_movies_inner.shape
```

#### 2.2. Full Join

```python
pd.merge(df1, df3, on='id', how='outer')
```

##### 2.2.1. Ejercicio

```python
pd.merge(df_movies, df_ratings, on='imdb_title_id', how='outer')
```

#### 2.3. Full Join Exclusivo

```python
df1.merge(df3, on='id', how='outer', indicator=True).query("_merge=='left_only' or _merge=='right_only'")
```

##### 2.3.1 Ejercicio

```python
df_movies.merge(df_ratings, on='imdb_title_id', 
                how='outer', 
                indicator=True).query("_merge=='left_only' or _merge=='right_only'")
```

#### 2.4 Left Join

```python
pd.merge(df1, df3, on='id', how='left')
```
